## 3. ELECCIÓN MÉTODO DE ESCALADO

### Enfoque 1: Evaluando alternativas

```python
inner_tss = TimeSeriesSplit(n_splits=3)

inner_scores = {}
```

```python
inner_tss
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline

# Alternativa 1: KNN con StandardScaler
pipeline_std = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsRegressor())
])

scores_std = -cross_val_score(pipeline_std, 
                              X_train, y_train, 
                              scoring='neg_root_mean_squared_error', 
                              cv=inner_tss)

inner_scores['KNN with StandardScaler'] = -scores_std.mean()

# Alternativa 2: KNN con MinMaxScaler
pipeline_min_max = Pipeline([
    ('scaler', MinMaxScaler()),
    ('knn', KNeighborsRegressor())
])

scores_min_max = -cross_val_score(pipeline_min_max, 
                              X_train, y_train, 
                              scoring='neg_root_mean_squared_error', 
                              cv=inner_tss)

inner_scores['KNN with MinMaxScaler'] = -scores_min_max.mean()

# Alternativa 3: KNN con RobustScaler
pipeline_robust = Pipeline([
    ('scaler', RobustScaler()),
    ('knn', KNeighborsRegressor())
])

scores_robust = -cross_val_score(pipeline_robust, 
                              X_train, y_train, 
                              scoring='neg_root_mean_squared_error', 
                              cv=inner_tss)

inner_scores['KNN with RobustScaler'] = -scores_robust.mean()

# Mostrar los resultados
for name, score in inner_scores.items():
    print(f"{name}: {score}")
```

### Enfoque 2: Evaluando alternativas con ajuste de hiper-parámetros

```python
from sklearn.model_selection import GridSearchCV

best_scores = {}

param_grid = {'knn__n_neighbors': [3, 5, 7, 9, 11, 13, 15]}

# Definir GridSearchCV para StandardScaler
grid_std = GridSearchCV(pipeline_std, 
                        param_grid, 
                        scoring='neg_root_mean_squared_error',
                        cv=inner_tss)

grid_std.fit(X_train, y_train)
best_scores['StandardScaler'] = -grid_std.best_score_

# Definir GridSearchCV para MinMaxScaler
grid_min_max = GridSearchCV(pipeline_min_max, 
                           param_grid, 
                           scoring='neg_root_mean_squared_error',
                           cv=inner_tss)

grid_min_max.fit(X_train, y_train)
best_scores['MinMaxScaler'] = -grid_min_max.best_score_

# Definir GridSearchCV para RobustScaler
grid_robust = GridSearchCV(pipeline_robust, 
                           param_grid, 
                           scoring='neg_root_mean_squared_error', 
                           cv=inner_tss)
                           
grid_robust.fit(X_train, y_train)
best_scores['RobustScaler'] = -grid_robust.best_score_

# Definir GridSearchCV para KNN sin scaler
param_grid_b = {'n_neighbors': [3, 5, 7, 9, 11, 13, 15]}
                           
grid_knn = GridSearchCV(KNeighborsRegressor(), 
                        param_grid_b, 
                        scoring='neg_root_mean_squared_error', 
                        cv=inner_tss)
                           
grid_knn.fit(X_train, y_train)
best_scores['KNN'] = -grid_knn.best_score_

# Mostrar los resultados
for name, score in best_scores.items():
    print(f"{name}: {score}")
```

## 4. EVALUACIÓN DE MODELOS

### A. EVALUACIÓN CON HIPERPARÁMETROS POR OMISIÓN

```python
from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.tree import DecisionTreeRegressor

from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet

from sklearn.svm import SVR

from sklearn.metrics import mean_squared_error

import time

scores = {}

"""MODELO 1: KNN"""
# Pipeline - modelo
pipeline_knn_std = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsRegressor())
])

#Entrenamiento y evaluación
time_start = time.time()
pipeline_knn_std.fit(X_train, y_train)
time_end = time.time()

total_time_trainning = time_end - time_start

knn_y_pred = pipeline_knn_std.predict(X_test)
knn_mse = mean_squared_error(y_test, knn_y_pred)

scores['KNN'] = {'MSE': knn_mse, 'Training Time': total_time_trainning}

"""MODELO 2: ÁRBOLES DE REGRESIÓN"""
# Modelo
regression_tree = DecisionTreeRegressor()

# Entrenamiento y evaluación
time_start = time.time()
regression_tree.fit(X_train, y_train)
time_end = time.time()

total_time_trainning = time_end - time_start

tree_y_pred = regression_tree.predict(X_test)

tree_mse = mean_squared_error(y_test, tree_y_pred)

scores['Decision Tree'] = {'MSE': tree_mse, 'Training Time': total_time_trainning}

"""MODELO 3: REGRESIÓN LINEAL NORMAL"""
# Modelo
linear_regression = LinearRegression()

# Entrenamiento y evaluación
time_start = time.time()
linear_regression.fit(X_train, y_train)
time_end = time.time()

total_time_trainning = time_end - time_start

linear_regression_y_pred = linear_regression.predict(X_test)

linear_regression_mse = mean_squared_error(y_test, linear_regression_y_pred)

scores['Linear Regression'] = {'MSE': linear_regression_mse, 'Training Time': total_time_trainning}

"""MODELO 4: LASSO"""
# Modelo
lasso_regression = Lasso()

# Entrenamiento y evaluación
time_start = time.time()
lasso_regression.fit(X_train, y_train)
time_end = time.time()

total_time_trainning = time_end - time_start

lasso_regression_y_pred = lasso_regression.predict(X_test)

lasso_regression_mse = mean_squared_error(y_test, lasso_regression_y_pred)

scores['Lasso Regression'] = {'MSE': lasso_regression_mse, 'Training Time': total_time_trainning}

"""MODELO 5: RIDGE"""
# Modelo
ridge_regression = Ridge()

# Entrenamiento y evaluación
time_start = time.time()
ridge_regression.fit(X_train, y_train)
time_end = time.time()

total_time_trainning = time_end - time_start

ridge_regression_y_pred = ridge_regression.predict(X_test)

ridge_regression_mse = mean_squared_error(y_test, ridge_regression_y_pred)

scores['Ridge Regression'] = {'MSE': ridge_regression_mse, 'Training Time': total_time_trainning}

"""MODELO 6: ELASTICNET"""
# Modelo 
elasticnet_regression = ElasticNet()

# Entrenamiento y evaluación
time_start = time.time()
elasticnet_regression.fit(X_train, y_train)
time_end = time.time()

total_time_trainning = time_end - time_start

elasticnet_regression_y_pred = elasticnet_regression.predict(X_test)
 
elasticnet_regression_mse = mean_squared_error(y_test, elasticnet_regression_y_pred)

scores['ElasticNet Regression'] = {'MSE': elasticnet_regression_mse, 'Training Time': total_time_trainning}

"""MODELO 7: SVM"""
# Modelo
pipeline_svm_std = Pipeline([
    ('scaler', StandardScaler()),
    ('svr', SVR())
])

#Entrenamiento y evaluación
time_start = time.time()
pipeline_svm_std.fit(X_train, y_train)
time_end = time.time()

total_time_trainning = time_end - time_start

svm_y_pred = pipeline_svm_std.predict(X_test)

svm_mse = mean_squared_error(y_test, svm_y_pred)

scores['SVM Regression'] = {'MSE': svm_mse, 'Training Time': total_time_trainning}
```

### B. EVALUACIÓN CON HPO

```python
scores_with_hpo = {}
inner_tss = TimeSeriesSplit(n_splits = 3)
```

```python
"""MODELO 1: KNN"""

# Hiper-parámetros
param_grid_knn = {'knn__n_neighbors': [3, 5, 7, 9, 11],
    'knn__weights': ['uniform', 'distance'],
    'knn__algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
}


pipeline_knn_std = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsRegressor())
])

# GridSearch
grid_search_knn = GridSearchCV(pipeline_knn_std,
                             param_grid,
                             scoring='neg_root_mean_squared_error',
                             cv=inner_tss)

# Entrenamiento
time_start = time.time()
grid_search_knn.fit(X_train, y_train)
time_end = time.time()

total_time_training_hpo = time_end - time_start

# Mejores hiper-parámetros
best_params = grid_search_knn.best_params_
best_model = grid_search_knn.best_estimator_

# Evaluación modelo
best_model_y_pred = best_model.predict(X_test)
best_model_mse = mean_squared_error(y_test, best_model_y_pred)

# Guardamos resultados
scores_with_hpo['KNN'] = {'MSE': best_model_mse, 
                                     'Training Time': total_time_training_hpo, 
                                     'Best Parameters': best_params}

"""MODELO 2: ÁRBOLES DE REGRESIÓN"""

param_grid_tree = {
    'tree__max_depth': [None, 10, 20, 30],
    'tree__min_samples_split': [2, 5, 10],
    'tree__min_samples_leaf': [1, 2, 4]
}

pipeline_tree_std = Pipeline([
    ('scaler', StandardScaler()),
    ('tree', DecisionTreeRegressor())
])

# GridSearch
grid_search_tree = GridSearchCV(pipeline_tree_std,
                                param_grid_tree,
                                scoring='neg_root_mean_squared_error',
                                cv=inner_tss)

# Entrenamiento
time_start = time.time()
grid_search_tree.fit(X_train, y_train)
time_end = time.time()

total_time_training_hpo = time_end - time_start

# Mejores hiper-parámetros
best_params = grid_search_tree.best_params_
best_model = grid_search_tree.best_estimator_

# Evaluación modelo
best_model_y_pred = best_model.predict(X_test)
best_model_mse = mean_squared_error(y_test, best_model_y_pred)

# Guardamos resultados
scores_with_hpo['Decision Tree'] = {'MSE': best_model_mse, 
                                               'Training Time': total_time_training_hpo, 
                                               'Best Parameters': best_params}
"""MODELO 3: REGRESIÓN LINEAL NORMAL"""

param_grid_lr = {
    'lr__fit_intercept': [True, False],
    'lr__copy_X': [True, False],
    'lr__n_jobs': [None, -1],
    'lr__positive': [False, True]
}

pipeline_lr_std = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LinearRegression())
])

# OGridSearch
grid_search_lr = GridSearchCV(pipeline_lr_std,
                              param_grid_lr,
                              scoring='neg_root_mean_squared_error',
                              cv=inner_tss)

# Entrenamiento
time_start = time.time()
grid_search_lr.fit(X_train, y_train)
time_end = time.time()

total_time_training_hpo = time_end - time_start

# Mejores hiper-parámetros
best_params = grid_search_lr.best_params_
best_model = grid_search_lr.best_estimator_

# Evaluación modelo
best_model_y_pred = best_model.predict(X_test)
best_model_mse = mean_squared_error(y_test, best_model_y_pred)

# Guardamos resultados
scores_with_hpo['Linear Regression'] = {'MSE': best_model_mse, 
                                                   'Training Time': total_time_training_hpo, 
                                                   'Best Parameters': best_params}

"""MODELO 4: LASSO"""

param_grid_lasso = {
    'lasso__alpha': [0.1, 0.5, 1.0, 2.0]
}

pipeline_lasso_std = Pipeline([
    ('scaler', StandardScaler()),
    ('lasso', Lasso())
])

# GridSearch
grid_search_lasso = GridSearchCV(pipeline_lasso_std,
                                  param_grid_lasso,
                                  scoring='neg_root_mean_squared_error',
                                  cv=inner_tss)

# Entrenamiento
time_start = time.time()
grid_search_lasso.fit(X_train, y_train)
time_end = time.time()

total_time_training_hpo = time_end - time_start

# Mejores hiper-parámetros
best_params = grid_search_lasso.best_params_
best_model = grid_search_lasso.best_estimator_

# Evaluación modelo
best_model_y_pred = best_model.predict(X_test)
best_model_mse = mean_squared_error(y_test, best_model_y_pred)

# Guardamos resultados
scores_with_hpo['Lasso Regression'] = {'MSE': best_model_mse, 
                                       'Training Time': total_time_training_hpo, 
                                       'Best Parameters': best_params}

"""MODELO 5: RIDGE"""

param_grid_ridge = {
    'ridge__alpha': [0.1, 0.5, 1.0, 2.0]
}

pipeline_ridge_std = Pipeline([
    ('scaler', StandardScaler()),
    ('ridge', Ridge())
])

# GridSearch
grid_search_ridge = GridSearchCV(pipeline_ridge_std,
                                 param_grid_ridge,
                                 scoring='neg_root_mean_squared_error',
                                 cv=inner_tss)

# Entrenamiento
time_start = time.time()
grid_search_ridge.fit(X_train, y_train)
time_end = time.time()

total_time_training_hpo = time_end - time_start

# Mejores hiper-parámetros
best_params = grid_search_ridge.best_params_
best_model = grid_search_ridge.best_estimator_

# Evaluación modelo
best_model_y_pred = best_model.predict(X_test)
best_model_mse = mean_squared_error(y_test, best_model_y_pred)

# Guardamos resultados
scores_with_hpo['Ridge Regression'] = {'MSE': best_model_mse, 
                                       'Training Time': total_time_training_hpo, 
                                       'Best Parameters': best_params}

"""MODELO 6: ELASTICNET"""

param_grid_elasticnet = {
    'elasticnet__alpha': [0.1, 0.5, 1.0, 2.0],
    'elasticnet__l1_ratio': [0.1, 0.3, 0.5, 0.7, 0.9]
}

pipeline_elasticnet_std = Pipeline([
    ('scaler', StandardScaler()),
    ('elasticnet', ElasticNet())
])

# GridSearch
grid_search_elasticnet = GridSearchCV(pipeline_elasticnet_std,
                                      param_grid_elasticnet,
                                      scoring='neg_root_mean_squared_error',
                                      cv=inner_tss)

# Entrenamiento
time_start = time.time()
grid_search_elasticnet.fit(X_train, y_train)
time_end = time.time()

total_time_training_hpo = time_end - time_start

# Mejores hiper-parámetros
best_params = grid_search_elasticnet.best_params_
best_model = grid_search_elasticnet.best_estimator_

# Evaluación modelo
best_model_y_pred = best_model.predict(X_test)
best_model_mse = mean_squared_error(y_test, best_model_y_pred)

# Guardamos resultados
scores_with_hpo['ElasticNet Regression'] = {'MSE': best_model_mse, 
                                             'Training Time': total_time_training_hpo, 
                                             'Best Parameters': best_params}
"""MODELO 7: SVM"""

param_grid_svr = {
    'svr__kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'svr__C': [0.1, 0.5, 1.0, 2.0],
    'svr__epsilon': [0.1, 0.5, 1.0, 2.0],
}

# Definición del pipeline
pipeline_svr_std = Pipeline([
    ('scaler', StandardScaler()),
    ('svr', SVR())
])

# Objetivo GridSearch
grid_search_svr = GridSearchCV(pipeline_svr_std,
                               param_grid_svr,
                               scoring='neg_root_mean_squared_error',
                               cv=inner_tss)

# Entrenamiento
time_start = time.time()
grid_search_svr.fit(X_train, y_train)
time_end = time.time()

total_time_training_hpo = time_end - time_start

# Mejores hiper-parámetros
best_params = grid_search_svr.best_params_
best_model = grid_search_svr.best_estimator_

# Evaluación modelo
best_model_y_pred = best_model.predict(X_test)
best_model_mse = mean_squared_error(y_test, best_model_y_pred)

# Guardamos resultados
scores_with_hpo['SVM Regression'] = {'MSE': best_model_mse, 
                                     'Training Time': total_time_training_hpo, 
                                     'Best Parameters': best_params}
```
